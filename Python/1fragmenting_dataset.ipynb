```python
import pandas as pd
```

✅ Step 1: Load the Dataset and Assign client_id

```python
#Loaded our dataset
df = pd.read_csv("d:\\kaggle\\bank-additional\\bank-additional\\bank-additional-full.csv")

# Assign unique client_id
df.insert(0,'client_id',range(1,len(df) + 1))

df.head()


```

✅ Step 2: Create the clients Table

```python
clients = df[['client_id', 'Age', 'Job', 'Marital', 'Education', 'Default', 'Housing', 'loan']]
clients.head()
```

✅ Step 3: Create the contacts Table

```python
contacts = df[['client_id', 'Contact', 'Month', 'Day_of_week', 'Duration']]
contacts.head()
```

✅ Step 4: Create the campaigns Table

```python
campaigns = df[['client_id', 'Campaign', 'Pdays', 'Previous', 'Poutcome']]
campaigns.head()
```

✅ Step 5: Create the economics Table

This table has various economic indicators

```python
economics = df[['client_id', 'Emp.var.rate', 'Cons.price.idx', 'Cons.conf.idx', 'Euribor3m', 'Nr.employed']]
economics.head()
```

✅ Step 6: Create the subscription Table (Target Variable)

```python
subscription = df[['client_id','Y']]
subscription.head()
```

✅ Step 7: Save These Tables as CSV

```python
clients.to_csv("clients.csv", index=False)
contacts.to_csv("contacts.csv", index=False)
campaigns.to_csv("campaigns.csv", index=False)
economics.to_csv("economics.csv", index=False)
subscription.to_csv("subscription.csv", index=False)
```
